```python
import pandas as pd

# Create a DataFrame with duplicates
data = {
    "ID": [101, 102, 103, 104, 102, 105, 106, 103],
    "Name": ["Alice", "Bob", "Charlie", "David", "Bob", "Eva", "Frank", "Charlie"],
    "Department": ["HR", "IT", "Finance", "Marketing", "IT", "HR", "Finance", "Finance"],
    "Salary": [5000, 6000, 7000, 8000, 6000, 5500, 7200, 7000]
}

df = pd.DataFrame(data)

print(df)
```

```python
df.duplicated() #duplicate True Otherwise False
```

```python
df 
```

```python
df.drop_duplicates(subset=['Department']) #it delete the last one manually
```

# handling null values

```python
import pandas as pd

# Create a Students DataFrame
data = {
    "Student_ID": [101, 102, 103, 104, 105, 106, 107, 108],
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva", "Frank", "Grace", "Hannah"],
    "Age": [20, 21, None, 22, 20, None, 23, 21],
    "Gender": ["Female", "Male", "Male", "Male", "Female", None, "Female", "Female"],
    "Department": ["CS", "IT", "CS", "SE", None, "IT", "SE", "CS"],
    "CGPA": [3.85, 3.60, None, 3.20, 3.95, 3.40, 3.75, 3.90],
    "Email": [
        "alice@example.com",
        "bob@example.com",
        "charlie@example.com",
        "david@example.com",
        None,
        None,
        "grace@example.com",
        "hannah@example.com"
    ]
}

df = pd.DataFrame(data)

print(df)
```

```python
df.dropna() #deletes all null value rows
```

```python
df.dropna(how='all',inplace=True) #if a rows contains null values all it will only delete that row
```

```python
df
```

```python
#colum wlise 
df.dropna(subset='Department')
```

```python
df.fillna(0) #null values will change to 0
```

```python
df['Department'] = df['Department'].fillna('unknown')
```

```python
df
```

```python
df['CGPA'] = df['CGPA'].fillna(df['CGPA'].mean())
```

```python
df
```

# apply function in df

```python
df
```

```python
#min max scalling
mn=df['Total_marks'].min()
print(mn)
mx=df['Total_marks'].max()
mx
```

# Scaled=(x-min)/(mx-mn)

```python
df['Scaled_marks']=df['Total_marks'].apply(lambda x:(x-mn)/(mx-mn))
```

```python
df
```

```python
#custom built in function 
def grading_system(marks):
    if marks >=260:
        return 'A'
    elif marks>=250:
        return 'A'
    else:
        return 'A-' #created custom built inn function 
df['Grade']=df['Total_marks'].apply(grading_system)
```

```python
df
```

```python
#all dataframe 
def marking_system(df):
    a=df['Data Structure Marks']*2
    b=df['Python Marks']*3
    c=df['Algorithm Marks']*4 
    return a+b+c

 
df['Excceptional_Marks']=df.apply(marking_system,axis=1)
```

```python
df
```
