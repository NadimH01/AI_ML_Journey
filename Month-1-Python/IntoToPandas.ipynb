# Transforming to dataframe from in built datas

```python
my_list=[['Nadim',23],['Munad',23],['Salman', 22]]
```

```python
list_df=pd.DataFrame(my_list,columns=['Name','Age'],index=[1,2,3])
print(type(list_df))
list_df
```

```python
my_tuple=(('Nadim',23),('salman',22),('Munad',24))
tuple_df=pd.DataFrame(my_tuple,columns=['Name','Age'])
tuple_df
```

```python
my_dict={
    'Name':['Nadim', 'Munad','salman'], 
    'Age':[23,23,22]
}
dict_df=pd.DataFrame(my_dict)
dict_df
```

```python
my_data=[
    {'Name': 'Nadim', 'Age':23, 'City':'Sylhet'},
    {'Name': 'Salman', 'Age':22, 'City':'Sylhet'},
    {'Name': 'Fahim', 'Age':23, 'City':'Barcelona'}
]
dat_df=pd.DataFrame(my_data)
dat_df
```
